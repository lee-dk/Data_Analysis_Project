### **자전거 대여소 정보**

```python
import pandas as pd
import numpy as np
```

```python
addr = pd.read_excel("data/공공자전거 대여소 정보.xlsx", sheet_name='대여소 현황')
print(addr.head())
addr.info() #총2083개의 데이터
```

```python
addr = addr.drop([0, 1, 2, 3], axis=0) # 4행까지 헤더 제거하기
addr = addr.drop(addr.columns[[6, 9]], axis=1) # 자전거 운영방식, 설치형태, 설치일자 제거
```

```python
addr.columns=['대여소번호', '대여소명', '위치(행정구)', '주소', '위도', '경도', 'LCD거치대수', 'QR거치대수'] # 직관적인 컬럼명으로 변경
addr.set_index('대여소번호', inplace=True)
addr.reset_index(inplace=True) # 행 번호 초기화(0부터 시작)
addr['대여소번호'] = addr['대여소번호'].astype(int) # 대여소번호의 형을 실수형에서 정수형으로 변환
display(addr.head(10))
```

```python
# 도로명 주소에서 구 주소로 변경하기 위해 주소 컬럼만 따로 저장한다.
df = addr['주소']
display(df)
df.to_excel('data/대여소주소.xlsx')
```

```python
new_addr= pd.read_excel("data/new_대여소주소.xlsx")
print(new_addr.head())
new_addr.info() # 총 2083개
```

```python
new_addr['소속동'] = new_addr['지번주소'].str.split(" ").str[2] # 동 정보만 뽑아냄.
display(new_addr.head(10)) # 앞 데이터 10개
display(new_addr.tail(10)) # 뒤 데이터 10개
new_addr = new_addr.drop(new_addr.columns[[0]], axis=1) # 도로명주소 제거
display(new_addr.head(5))
```

```python
# 새로 만든 동이 포함된 주소를 기존 addr에 붙이기
addr.insert(4, '지번주소', new_addr['지번주소'])
addr.insert(5, '소속동', new_addr['소속동'])
display(addr.head())
```

```python
addr.info()
addr.head()
```

```python
# 기존에 갖고있던 도로명 주소 칼럼 삭제하기
addr = addr.drop(addr.columns[[3]], axis=1)
```

```python
# 자전거 대여소의 동 주소 뽑아내기 완료.
addr.head()
```

```python
addr.set_index('소속동', inplace=True)
```

```python
# LCD 거치대수와 QR 거치대수의 결측치 값 0으로 변경
addr.fillna(0, inplace=True)
display(addr.head())
```

```python
#거치대수를 합치고, 기존에 있는 LCD, QR거치대수 칼럼을 제거한다.
addr['총거치대수'] = addr[['LCD거치대수', 'QR거치대수']].sum(axis=1)
```

```python
addr = addr.drop(addr.columns[[6]], axis=1)
```

```python
addr = addr.drop(addr.columns[[6]], axis=1)
display(addr.head())
```

```python
addr.info() # 총 2083개의 칼럼
```

```python
cnt = addr.groupby(addr.index)['총거치대수'].sum()
display(cnt.sort_values(ascending=False).head()) # 거치대수가 많은 순서대로 내림차순 정렬하였음.
```
